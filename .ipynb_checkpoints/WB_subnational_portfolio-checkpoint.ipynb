# Subnational Portfolio Identification of the World Bank projects
### Background 
The IEG has started a new major evaluation on managing urban spatial growth. Despite the nature of the spatial focus, there is no systematic approach to identify subnational portoflio. In this excercise, I used `geoid` provided to each project to identify the location of the sites.

### What is Geoid?
Geoid is the unique geographical location ID, which is managed by geonames.org. In the World Bank, upon the approval of the project, the ITS reviews the relevant project document (PAD, EIA, etc), and provides the geoid.<br>

### Objective
to identify subnational level project portfolio in the Urban Spatial Growth Evaluation

### Methodology
**Step 1. Import and clean data**: (1) import global portfolio with GeoID, (2) import project portfolio, (3) merge global  and project portfolio.<br>
**Step 2. Read geoid and detect locations**: convert unique geo_id per project and develop a geodataframe. Use `geocoder` library.<br>
**Step 3. Visualize subnational portfolio**: Visualize the subnational portfolio.<br>
**Step 4. Observe correlation between the choice of sites and population.** (1) extract subnational population, (2) merge pop data to subnational portfolio, and (3) observe correlation between the numbe of sites and population.

Processing for GeoLocID is necessary from the following points.
- GeoLocID has numerous entries devided by semi-colon. Need splitting.
- It needs to be replaed into the list for next processing.
- GeoLocID has forward zeroes, which are not true id. Remove forward zeroes to enable geocoder to read.

```python
# Split by semi-colon, make a list of GeoLocID, and remove forward zeroes.
ids_lst =[]
cost_lst = []
geo_lst = []

for index, row in WB_data.iterrows():
    i_d = row['id']
    cost = row['lendprojectcost']
    geoid = row['GeoLocID']
    ids_lst.append(i_d)
    cost_lst.append(cost)
    geo_lst.append(geoid.split(';'))
    
WB_geodata = pd.DataFrame({'P_number':ids_lst,
                           'Project_Cost':cost_lst,
                           'GeoLocID':geo_lst})
WB_geodata['GeoLocID'] = WB_geodata['GeoLocID'].apply(lambda x: [int(y) for y in x]) # this removes forward zeroes.
WB_geodata = WB_geodata.set_index('P_number')
```

## Step 2. Read geoid and detect locations
- Basic script: g = geocoder.geonames(6094817, method='details', key='USERNAME')
- Example: g.lat; g.lng; g.geonames_id; g.address; g.feature_class; g.class_description; g.code; g.description; g.continent; g.country_geonames_id; g.country_code; g.country; g.state; g.state_code; g.state_geonames_id; g.admin2; g.admin3; g.admin4; g.admin5; g.population; g.srtm3; g.bbox
- Remarks: `geocoder` allows 1000 pull requests per hour for free of charge.

```python
import geocoder
import time # for automatic computation of geocoder
```

First, the `sn_portfolio` dataframe has numerous data entry in each GeoLocID, which needs to be split into single GeoLocID in each row.

```python
# Split the geoloc list
row_bag, id_bag = [], []
shorty = sn_portfolio[:]
for index, row in shorty.iterrows():
    bag_of_geolocs = row.GeoLocID
    for individual_geoloc in bag_of_geolocs:
        row_bag.append(row)
        id_bag.append(individual_geoloc)
split_df = pd.DataFrame(row_bag)
split_df['ACTUAL_geoloc'] = id_bag
split_df = split_df.drop('GeoLocID', axis = 1)
split_df = split_df.reset_index()
split_df['org_id'] = split_df.index
split_df.head()
```

Next, we will define a function that converts `GeoID` into place names.<br>
You need to create your account at Geonames.org, which can be accessed at https://www.geonames.org/login.<br>
After you create your account, you need to enable the free web services for allowing geocoder to pull data.<br>
Go to http://www.geonames.org/manageaccount, and click `Click here to enable` in the page.

```python
USERNAME = 'YOUR_ACCOUNT' # Replace YOUR_ACCOUNT
```

```python
# Define function that converts GeoID into place names.
def get_location_wb(place_ID):
    state = geocoder.geonames(place_ID, method = 'details', key = USERNAME).state
    admin2 = geocoder.geonames(place_ID, method = 'details', key = USERNAME).admin2 
    admin3 = geocoder.geonames(place_ID, method = 'details', key = USERNAME).admin3
    admin4 = geocoder.geonames(place_ID, method = 'details', key = USERNAME).admin4 
    admin5 = geocoder.geonames(place_ID, method = 'details', key = USERNAME).admin5
    population = geocoder.geonames(place_ID, method = 'details', key = USERNAME).population
    bbox = geocoder.geonames(place_ID, method = 'details', key = USERNAME).bbox
    lat = geocoder.geonames(place_ID, method = 'details', key = USERNAME).lat
    lng = geocoder.geonames(place_ID, method = 'details', key = USERNAME).lng
    return state, admin2, admin3, admin4, admin5, population, bbox, lat, lng
```

Then, with reference to `split_df`, the following function will pull place names. However, geocoder license only allows **1,000 pull requests per hour**. Thus, the function needs 1 hour rest after pulling 1,000 items.<br>
Total required time is **52 hours**. (5,711 sites x 9 variables / 1,000)

```python
counter = 0
```

```python
# Convert geoloc id to location names
results_dict = {}
for index, row in split_df.iterrows():
    calls = 9
    counter += calls
    if counter < 1000:
        print('downloading for: %s' % index)
        state, admin2, admin3, admin4, admin5, population, bbox, lat, lng = get_location_wb(row.ACTUAL_geoloc)
        results = [state, admin2, admin3, admin4, admin5, population, bbox, lat, lng]
        results_dict[index] = results
    else:
        print('calls is going to exceed 1,000 per hour limit. sleeping!')
        time.sleep(60 * 60)
        counter = 0
    print('calls remaining this hour: %s' % (1000 - counter)) 
    
cols = ['state', 'admin2', 'admin3', 'admin4', 'admin5', 'population', 'bbox', 'lat', 'lng']
geo_coding_results = pd.DataFrame.from_dict(results_dict, orient = 'index', columns = cols)
```

```python
# to merge the dataframe later, add index into the column.
geo_coding_results['org_id'] = geo_coding_results.index
```

```python
# Save file.
geo_coding_results.to_csv('geo_coding_results.csv')
```

## Step 3. Visualize subnational portfolio
Now we have the dataframe. Each row represents the level of intervention. I need to develop subnational portfolio maps of state level and admin2 level. I will use latitude and longitude data to spatially join the dataframe and admin boundary shapefile. Initially, I tried to match the administraton name (state, admin2) between the dataframe and admin boundary shapefile, however it did not work as a lot of variations in the administration name were identified. Alternativel, I used the point data of the location with reference to state and admin2.
<img src="image\image_spatial_join.png" width = '40%'>

```python
import geopandas as gpd
import shapely.geometry as shp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Convert latitude and longitude into geometry.
portfolio['geometry'] = portfolio[['lng','lat']].apply(shp.Point, axis=1)
portfolio = portfolio.drop(['lat', 'lng'], axis = 1)
crs = {'init': 'epsg:4326'}
portfolio = gpd.GeoDataFrame(portfolio, crs=crs)
print(type(portfolio), portfolio.crs)
portfolio.tail()
```

Read administration bounary shapefiles.

```python
country_boundary = gpd.read_file(r".\shapefile\WB_ADM0.shp")
state_boundary = gpd.read_file(r".\shapefile\WB_ADM1.shp")
admin2_boundary = gpd.read_file(r".\shapefile\WB_ADM2.shp")
capital = gpd.read_file(r".\shapefile\Nat_Capitals2.shp")
```

```python
state_boundary.head(2)
```

```python
admin2_boundary.head(2)
```

Execute spatial join of shapefiles and dataframe, so that we can identify the project locations in each administrative level.

```python
# Make dataframe spatially joined by country.
country_geom_portfolio = gpd.sjoin(country_boundary, portfolio, op='contains', how='left')
country_geom_portfolio = country_geom_portfolio.dropna(subset = ['Country']).reset_index()
country_geom_portfolio.shape
```

```python
# Make dataframe spatially joined by state.
state_geom_portfolio = gpd.sjoin(state_boundary, portfolio, op='contains', how='left')
state_geom_portfolio = state_geom_portfolio.dropna(subset = ['state']).reset_index()
state_geom_portfolio.shape
```

```python
# Make dataframe spatially joined by admin2.
admin2_geom_portfolio = gpd.sjoin(admin2_boundary, portfolio, op='contains', how='left')
admin2_geom_portfolio = admin2_geom_portfolio.dropna(subset = ['admin2']).reset_index()
admin2_geom_portfolio.shape
```

Make portfolio mapping. (1) Country, (2) State, (3) ADM2

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, ncols = 1, figsize = (15, 15))

# Country
ax1 = country_boundary.plot(ax = ax1, facecolor = 'none', edgecolor = 'gray', linewidth = .5)
country_geom_portfolio.plot(ax = ax1, color = 'skyblue', edgecolor='black', linewidth = .3)
ax1.set_title('MUSG Country Portfolio')
ax1.set_axis_off()

# State
ax2 = country_boundary.plot(ax = ax2, facecolor = 'none', edgecolor = 'gray', linewidth = .5)
state_geom_portfolio.plot(ax = ax2, color = 'skyblue', edgecolor='black', linewidth = .3)
ax2.set_title('MUSG State Portfolio')
ax2.set_axis_off()

# ADM2
ax3 = country_boundary.plot(ax = ax3, facecolor = 'none', edgecolor = 'gray', linewidth = .5)
admin2_geom_portfolio.plot(ax = ax3, color = 'skyblue', edgecolor='black', linewidth = .3)
ax3.set_title('MUSG Admin2 Portfolio')
ax3.set_axis_off()
```

#### Count the number of projects and approximate cumulative cost by admin boundary

State: Number of Sites and Cumulative Cost

```python
# 1 State: Number of Sites
state_geom_portfolio_grouped = state_geom_portfolio.groupby(['Region', 'ADM0_NAME','ADM1_NAME']).agg({'P_number':'count'}).reset_index()
state_geom_portfolio_grouped.columns = ['Region', 'ADM0_NAME','ADM1_NAME','N_Sites']
state_geom_portfolio_grouped = pd.merge(state_geom_portfolio_grouped, state_boundary[['ADM0_NAME','ADM1_NAME', 'geometry']], on = ['ADM0_NAME','ADM1_NAME'], how = 'inner')
state_geom_portfolio_grouped = gpd.GeoDataFrame(state_geom_portfolio_grouped)
print(state_geom_portfolio_grouped.shape, type(state_geom_portfolio_grouped))
state_geom_portfolio_grouped.head()
```

```python
# 2 State: Cost per Site
state_geom_portfolio_grouped_cost = state_geom_portfolio.groupby(['Region', 'ADM0_NAME','ADM1_NAME']).agg({'cost_per_site':'sum'}).reset_index()
state_geom_portfolio_grouped_cost.columns = ['Region', 'ADM0_NAME','ADM1_NAME','Cumulative_Cost']
state_geom_portfolio_grouped_cost = pd.merge(state_geom_portfolio_grouped_cost, state_boundary[['ADM0_NAME','ADM1_NAME', 'geometry']], on = ['ADM0_NAME','ADM1_NAME'], how = 'inner')
state_geom_portfolio_grouped_cost = gpd.GeoDataFrame(state_geom_portfolio_grouped_cost)
print(state_geom_portfolio_grouped_cost.shape, type(state_geom_portfolio_grouped_cost))
state_geom_portfolio_grouped_cost.head()
```

```python
# 3 State: Merge N_Sites and Cost
state_geom_portfolio_cost = pd.merge(state_geom_portfolio_grouped, state_geom_portfolio_grouped_cost[['ADM0_NAME', 'ADM1_NAME', 'Cumulative_Cost']], on = ['ADM0_NAME', 'ADM1_NAME']).reset_index()
state_geom_portfolio_cost = gpd.GeoDataFrame(state_geom_portfolio_cost)
state_geom_portfolio_cost.head()
```

ADM2: Number of Sites and Cumulative Cost

```python
# 1 Admin2: Number of Sites
admin2_geom_portfolio_grouped = admin2_geom_portfolio.groupby(['Region', 'ADM0_NAME','ADM1_NAME', 'ADM2_NAME']).agg({'P_number':'count'}).reset_index()
admin2_geom_portfolio_grouped.columns = ['Region', 'ADM0_NAME','ADM1_NAME', 'ADM2_NAME', 'N_Sites']
admin2_geom_portfolio_grouped = pd.merge(admin2_geom_portfolio_grouped, admin2_boundary[['ADM0_NAME','ADM1_NAME', 'ADM2_NAME', 'geometry']], on = ['ADM0_NAME','ADM1_NAME', 'ADM2_NAME'], how = 'inner')
admin2_geom_portfolio_grouped = gpd.GeoDataFrame(admin2_geom_portfolio_grouped)
print(admin2_geom_portfolio_grouped.shape, type(admin2_geom_portfolio_grouped))
admin2_geom_portfolio_grouped.head()
```

```python
# 2 Admin2: Cost per Site
admin2_geom_portfolio_grouped_cost = admin2_geom_portfolio.groupby(['Region', 'ADM0_NAME','ADM1_NAME', 'ADM2_NAME']).agg({'cost_per_site':'sum'}).reset_index()
admin2_geom_portfolio_grouped_cost.columns = ['Region', 'ADM0_NAME','ADM1_NAME', 'ADM2_NAME', 'Cumulative_Cost']
admin2_geom_portfolio_grouped_cost = pd.merge(admin2_geom_portfolio_grouped_cost, admin2_boundary[['ADM0_NAME','ADM1_NAME', 'ADM2_NAME', 'geometry']], on = ['ADM0_NAME','ADM1_NAME', 'ADM2_NAME'], how = 'inner')
admin2_geom_portfolio_grouped_cost = gpd.GeoDataFrame(admin2_geom_portfolio_grouped_cost)
print(admin2_geom_portfolio_grouped_cost.shape, type(admin2_geom_portfolio_grouped_cost))
admin2_geom_portfolio_grouped_cost.head()
```

```python
# 3 Admin2: Merge N_Sites and Cost
admin2_geom_portfolio_cost = pd.merge(admin2_geom_portfolio_grouped, admin2_geom_portfolio_grouped_cost[['ADM0_NAME', 'ADM1_NAME', 'ADM2_NAME', 'Cumulative_Cost']], on = ['ADM0_NAME', 'ADM1_NAME', 'ADM2_NAME']).reset_index()
admin2_geom_portfolio_cost = gpd.GeoDataFrame(admin2_geom_portfolio_cost)
admin2_geom_portfolio_cost.head()
```

## Step 4. Extract subnational population and merge pop data to subnational portfolio
In this step, I am interested in correlation factors that would determine the location of the project. In this excercise, I am going to check if subnational location and population are correlated.

```python
import rasterio
from rasterio.merge import merge
from rasterio.plot import show
from rasterstats import zonal_stats
import glob
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.environ['PROJ_LIB'] = r"C:\\WBG\Anaconda\Library\share"
```

Here, I will use 2015 World Pop 1km2 grid data in Asia, LAC, and AFR, which are available at https://www.worldpop.org/geodata/listing?id=17.<br>
**First**, I will merge these datasets into one single raster data.<br>
**Second**, I will dump population dataset into state and ADM2 geodataframe to provide population of the admin boundary.<br>
**Third**, I will check the correlation between location and population.<br><br>
**Step 1: Merge separate worldpop datasets**

```python
# File and folder paths
dirpath = r".\raster_data"
out_fp = r".\world_pop_merged.tif"

# Make a search criteria to select the raster files
search_criteria = "*2015*.tif"

q = os.path.join(dirpath, search_criteria)
print(q)
```

```python
dem_fps = glob.glob(q)
dem_fps
```

```python
world_pop_files_to_mosaic = []
for fp in dem_fps:
    world_pop = rasterio.open(fp)
    world_pop_files_to_mosaic.append(world_pop)

world_pop_files_to_mosaic
```

```python
mosaic, out_trans = merge(world_pop_files_to_mosaic)
```

```python
show(mosaic, cmap='OrRd')
```

```python
out_meta = world_pop.meta.copy()
```

```python
out_meta.update({"driver": "GTiff",
                 "height": mosaic.shape[1],
                 "width": mosaic.shape[2],
                 "transform": out_trans,
                 "crs": "+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs"
                })
```

```python
with rasterio.open(out_fp, "w", **out_meta) as dest:
    dest.write(mosaic)
```

# Scatter plot of Corrleation
### State Level

```python
# Scatter plot - State Level    
sns.set(style="whitegrid")

f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(14, 5))
fig.suptitle('State level Correlation')
sns.despine(f, left=True, bottom=True)

ax1 = sns.scatterplot(x="Cumulative_Cost", 
                y="population",
                hue="Region",
                palette="ch:r=-.2,d=.3_r",
                linewidth=0,
                data=state_pop_portfolio_no_zero, 
                ax=ax1)
ax1.set_title('Cumulative Cost and Population - State')

ax2 = sns.scatterplot(x="N_Sites", 
                y="population",
                hue="Region", 
                palette="ch:r=-.2,d=.3_r",
                linewidth=0,
                data=state_pop_portfolio_no_zero, 
                ax=ax2)
ax2.set_title('Number of Sites and Population - State')
```

```python
# Correlation with Population - State Level
cost_corr = state_pop_portfolio_no_zero['Cumulative_Cost'].corr(state_pop_portfolio_no_zero['population'])
sites_corr = state_pop_portfolio_no_zero['N_Sites'].corr(state_pop_portfolio_no_zero['population'])
print('Correlation between cost and population is %.2f' % cost_corr)
print('Correlation between number of sites and population is %.2f' % sites_corr)
```
